**Exercise 4**

a

```python
def Q1(l):
    return np.percentile(l, 25)
```

```python
def Q2(l):
    return np.percentile(l, 50)
```

```python
def Q3(l):
    return np.percentile(l, 75)
```

b

```python
dataw = [76, 72, 88, 60, 72, 68, 80, 64, 68, 68, 80, 76, 68, 72, 96, 72, 68, 72, 64, 80,
64, 80, 76, 76, 76, 80, 104, 88, 60, 76, 72, 72, 88, 80, 60, 72, 88, 88, 124, 64]
```

```python
Q1(dataw)
```

```python
Q2(dataw)
```

```python
Q3(dataw)
```

c

```python
statistics.median(dataw)
```

e

```python
def EI(l):
    return Q3(l) - Q1(l)
```

```python
EI(dataw)
```

```python
def ab1(l):
    return Q1(l) - 1.5*EI(l)
```

```python
def ab2(l):
    return Q3(l) + 1.5*EI(l)
```

```python
def val_ab(l):
    result = []
    for val in l:
        if val < ab1(l) or val > ab2(l):
            result.append(val)
    return result
```

f

```python
ab1(dataw)
```

```python
ab2(dataw)
```

```python
val_ab(dataw)
```
